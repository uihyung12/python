```python
def run(dirname):
    try:
        # 입력으로 들어온 디렉토리 내의 파일과 서브 디렉토리 리스트를 구함
        filenames = os.listdir(dirname)
        # 각 디렉토리와 파일을 하나씩 순회하면서...
        for filename in filenames:
            # 디렉토리 or 파일의 전체 경로로 묶는다.
            full_filename = os.path.join(dirname, filename)
            # 만약 디렉토리일 경우에 서브 디렉토리에 대해 재귀함수로 반복 작업을 수행한다.
            if os.path.isdir(full_filename):
                run(full_filename)
            else:
                # 파일일 경우 txt 파일인지 확인하여 변환 작업을 한다.
                ext = os.path.splitext(full_filename)[-1]
                if ext == '.txt':
                    #print(full_filename)
                    convert(full_filename)
    except PermissionError:
        pass

def convert(filename):
    label = ''
    # 입력으로 들어온 파일을 연다.
    f1 = open(filename, 'r', encoding="UTF-8")
    while True:
        # 파일의 내용을 한 줄 씩 읽어온다.
        line = f1.readline()
        if not line:
            break
        # '\n'은 제거하고 공백을 기준으로 내용을 나눠준다.
        label_split = line.rstrip('\n').split(' ')
#         print(line)
#         print(label_split)
        # 첫 번째 항목은 타입이므로 이것이 1일 경우 0으로 치환한다.
#         print(filename)
#         print(label_split)
        
#         if label_split[0] == '3':
#             label_split[0] = '0'
#             label = label + label_split[0]+' '+label_split[1]+' '+label_split[2]+' '+label_split[3]+' '+label_split[4]+"\n"
#         if label_split[0] == '1':
#             label = label + label_split[0]+' '+label_split[1]+' '+label_split[2]+' '+label_split[3]+' '+label_split[4]+"\n"
#         if label_split[0] == '0':
#             label_split[0] = '2'
#         elif label_split[0] == '1':
#             label_split[0] = '5'
#         elif label_split[0] == '2':
#             label_split[0] = '6'
#         elif label_split[0] == '3':
#             label_split[0] = '4'
#         elif label_split[0] == '4':
#             label_split[0] = '3'
#         elif label_split[0] == '5':
#             label_split[0] = '1'
#         elif label_split[0] == '6':
#             label_split[0] = '0'
        print(f1)
#         #normalization
#         a = float('{}'.format(label_split[3]))/float('{}'.format(label_split[0]))
#         b = float('{}'.format(label_split[4]))/float('{}'.format(label_split[1]))
#         c = float('{}'.format(label_split[5]))/float('{}'.format(label_split[0]))
#         d = float('{}'.format(label_split[6]))/float('{}'.format(label_split[1]))
        a = float('{}'.format(label_split[1]))*float('{}'.format(label_split[5]))
        b = float('{}'.format(label_split[2]))*float('{}'.format(label_split[6]))
        c = float('{}'.format(label_split[3]))*float('{}'.format(label_split[5]))
        d = float('{}'.format(label_split[4]))*float('{}'.format(label_split[6]))
        
# #         a = float(label_split[3]) / float(label_split[0])
# #         b = float(label_split[4]) / float(label_split[1])
# #         c = float(label_split[5]) / float(label_split[0])
# #         d = float(label_split[6]) / float(label_split[1])
# # #         a = float(a)
# # #         print(type(a))
    
        a = str(round(a,6))
        b = str(round(b,6))
        c = str(round(c,6))
        d = str(round(d,6))
    
    
#         # 치환된 내용을 포함하여 모든 항목들을 모아서 한 줄로 만들어 문자열 변수에 쌓는다.
        label = label+label_split[0] +' '+ a + ' '+b+' '+c+' '+d+' '+'\n'
    f1.close();

    # 새로 저장될 label 파일의 경로
    new_filename = filename.replace('qwe', 'asd')

    # 새로운 디렉토리 생성(질문)
    dir = new_filename.replace('\\', '/').split('/')
    new_path = dir[0] + "/"
    for i in range(1, len(dir)-1):
        new_path = os.path.join(new_path, dir[i])
        if not os.path.isdir(new_path):
            os.makedirs(os.path.join(new_path))

    # 새로운 label을 파일에 기록한다.
    with open(new_filename, 'w') as f2:
        f2.write(label)
```

```python
import os
import re
```

```python
def blank_txt(dirname):
    try:
        files = os.listdir(dirname)#file들의 list 불러오기
        
        for file in files:
            full_path = os.path.join(dirname, file)#풀경로를 합치기
            
            
            if os.path.isdir(full_path):#만약에 파일이 있다면 다시 검색하기(재귀법)
                blank_txt(full_path)
                
                
            else:
                ext = os.path.splitext(full_path)[-1]#파일형식이 어떤 것인지 확인하기
                txt_list = []#txt 파일들만 불러오기
                
                if ext == '.txt':
                    
                    file_path = full_path.replace("\\",'/')#full_path
                    txt_list.append(file_path)
                    
                for i in txt_list:
                    if os.path.getsize(i) > 0:#내용이 있는 경우
                        print('confirm file: {}'.format(i))
                        
                    else:
                        
                        print('empty file : {}'.format(i))#내용이 없는 경우
                        os.remove(i)
                    
                    
                    
                    
                    
        
        
        
    except:
        pass
```

```python
def blank_remove(dirname):
    files = os.listdir(dirname)
    
    for file in files:
        full_path = os.path.join(dirname, file)
        
        if os.path.isdir(full_path):
            blank_remove(full_path)
            
        else:
            replace_path = full_path.replace("\\",'/')
            
            split_path = replace_path.split('/')[-1]
#             print(split_path)
            
            replace_blank = split_path.replace(' ','_')
#             print(replace_blank)
            os.rename('D:/Data_collection/car_data/car_load_data/topcar_except/weekly/101547 (5-20-2020 4-06-58 PM)/'+split_path,'D:/Data_collection/car_data/car_load_data/topcar_except/weekly/101547 (5-20-2020 4-06-58 PM)/'+replace_blank)
```

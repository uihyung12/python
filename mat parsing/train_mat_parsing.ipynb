```python
def file_txt(dirname):#.jpg.txt 되어있는 파일명을 .txt로 바꾸기
    try:
        txt_lists = os.listdir(dirname)
        
        for file in txt_lists:
            full_path = os.path.join(dirname, file)
            
            if os.path.isdir(full_path):
                file_txt(full_path)
                
            else:
                
                ext = os.path.splitext(full_path)[-1]
                
                if ext == '.txt':
                    print(ext)
#                     patten = '.jpg'
#                     repl = ''
#                     text = '{}'.format(file)
                    
#                     m = re.sub(pattern= patten, repl=repl, string = text)
                    
#                     os.rename('D:/train_mat_parsing/'+file, 'D:/train_mat_parsing/'+m)
                
        
        
    except PermissionError:
        pass
```

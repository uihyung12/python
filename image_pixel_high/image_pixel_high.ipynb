```python
def adjust_gamma(image, gamma=1.0):
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")

    return cv2.LUT(image, table)
```

```python
def image_pixel(dirname):
    try:
        files = os.listdir(dirname)
        
        for file in files:
        
            full_image = os.path.join(dirname, file)
            
            if os.path.isdir(full_image):
                image_pixel(full_image)
                
            else:
#                 full_image = full_image.replace('\\','/')
#                 print(full_image)
                ext = os.path.splitext(full_image)[-1]
                if ext == '.jpg':
            
                    img = cv2.imread(full_image)
                    print(full_image)
                    print(img)
                    gamma = 2.0

                    adjusted = adjust_gamma(img, gamma=gamma)
                    cv2.imwrite(full_image.split('.')[0]+'.jpg',adjusted)
        
        
        
    except PermissionError:
        pass
```

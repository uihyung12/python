```python
import numpy as np
import cv2
import os
```

```python
def jpg_convert(dirname):#전체적 파일 불러요기
    list_jpg = os.listdir(dirname)
    
    for file in list_jpg:
        full_file = os.path.join(dirname, file)
        
        ext = os.path.splitext(full_file)[-1]
        
        if ext == '.jpg':
            transform(full_file)
        
        




def transform(filename):
    dirname = filename.replace("\\",'/')#파일명이 \ 나오기 때문에 / 로 바꾼다

    
    
    img = cv2.imread(dirname)
#     print(img)
#     print(filename)
    h, w = img.shape[:2]
#     new_filename = filename.replace('right_rotate','right')
    
#     M1 = cv2.getRotationMatrix2D((w/2, h/2), 45, 1)
    M2 = cv2.getRotationMatrix2D((w/2, h/2), 180, 1)#180도 변환
    
#     img2 = cv2.warpAffine(img, M1, (w, h))
    img3 = cv2.warpAffine(img, M2, (w, h))
    cv2.imwrite('D:/Data_collection/180/{}'.format(dirname.split('/')[-1]), img3)
    
#     cv2.imshow('45-Rotate', img2)
#     cv2.imshow('90-Rotate', img3)
    
    
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
```

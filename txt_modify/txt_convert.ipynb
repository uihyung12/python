```python
import os
```

```python
def txt_convert(dirname):
    try:
        files = os.listdir(dirname)

        for file in files:
            full_file = os.path.join(dirname, file)

            if os.path.isdir(full_file):
                txt_convert(full_file)

            else:
                ext = os.path.splitext(full_file)[-1]

                if ext == '.txt':
                    convert(full_file)
    except PermissionError:
        pass
    
    
    
def convert(filename):
    label = ''
    
    f1 = open(filename, 'r')
    while True:
        
        line = f1.readline()
        if not line:
            break
            
        label_split = line.rstrip('\n').split(' ')
#         print(label_split)
        
        if label_split[0] == '2':#차
            label_split[0] = '0'
            label = label + label_split[0] + ' ' + label_split[1] + ' ' + label_split[2] + ' ' + label_split[3] + ' ' + label_split[4]+'\n'
        elif label_split[0] == '5':#버스
            label_split[0] = '4'
            label = label + label_split[0] + ' ' + label_split[1] + ' ' + label_split[2] + ' ' + label_split[3] + ' ' + label_split[4]+'\n'
        elif label_split[0] == '1':#트럭
#             label_split[0] = '1'
            label = label + label_split[0] + ' ' + label_split[1] + ' ' + label_split[2] + ' ' + label_split[3] + ' ' + label_split[4]+'\n'
#         elif label_split[0] == '3':
#             label = label + label_split[0] + ' ' + label_split[1] + ' ' + label_split[2] + ' ' + label_split[3] + ' ' + label_split[4]+'\n'
#         elif label_split[0] == '4':
#             label = label + label_split[0] + ' ' + label_split[1] + ' ' + label_split[2] + ' ' + label_split[3] + ' ' + label_split[4]+'\n'


            
    f1.close()
        
        
#     f1.close():
        
    new_filename = filename.replace('night', 'night_copy')
    
    dir = new_filename.replace("\\",'/').split('/')
    print(dir)
    print(len(dir))
    new_path = dir[0] + "/"
    
#     print(len(new_path))
    for i in range(1, len(dir)-1):
        new_path = os.path.join(new_path, dir[i])
        if not os.path.isdir(new_path):
            os.makedirs(os.path.join(new_path))

    # 새로운 label을 파일에 기록한다.
    with open(new_filename, 'w') as f2:
        f2.write(label)
        
        
    
```

#### 모든 if 제어문에 대한 고찰 

if (조건) :      #(조건)이 참이면 = (조건)의 return값이 1(True)이면  
    (종속절)     # (종속절) 실행된다.

- (조건) -> 논리/비교연산 : 결과가 True = return값이 1(True)
- (조건) -> 숫자 자체 : 0이 아닌 숫자 = return값이 1(True) ★
- (조건) -> 문자 자체 : ' '이 아닌 문자 = return값이 1(True) ★

```python
if 16:  #0 도 해보기 
    print("너 0이 아니구나! True")
```

```python
if '':  #'홍'도 해보기 
    print("너 ''이 아니구나! True")
else:
    print("너 공백이구나! False")
```

### for문에 대한 고찰  

for i in range(시작, 끝+1, 간격):  
    (종속절)

i가 (시작)부터 (끝)까지 도달할 때까지 (종속절)을 반복   

★ 3개 암기
- 시작: 쓰지 않아도 됨. 대신 시작= 0으로 간주 
- 끝+1: 무조건 써야함.
- 간격: 쓰지 않아도 됨. 대신 간격=1로 간주

```python
for i in range(100) : #시작=0, 끝+1=100(끝=99), 간격=1
    print(i, end=' ')
```

### 무한반복 + continue/break 사용하기  

```python
number = 0
while 1:
    print(number, end=' ')
    number += 1
    
    if number == 10:
        break
```

### 0528 수업

```python
number = 0
while 1:
    print(number, end =' ')
    number += 1
    
    if number < 10:
        continue   #현재시행 여기서 끝내고 다시 while문 처음으로 감
    else:
        break      # 반복시행(while문) 종료함
        
    print('-')    #그래서 실행안됨
```

Q. data 안에 5가 들어있는지 확인하고, 5가 있다면 "찾았다!" 출력하세요

Q. 2개의 리스트를 동시에 for문으로 제어할 수 있을까?

#### zip 

```python
name = ['이원재', '김가영', '김강빈', '김민철', '김원중', '김홍덕']
hobby = ['song','movie','TV','coding','sleep','cycle']

#for n,h in name, hobby:  #에러! 
for n,h in zip(name, hobby):  #해결!
    print(n,h)
```

#### 다중 for문 

```python
data = [['이원재', '김가영', '김강빈', '김민철', '김원중', '김홍덕'],
        ['song','movie','TV','coding','sleep','cycle']]
for d in data:
    print(d)
```

Q. 그런데, data의 각 원소 하나하나를 지정해야 된다면?

```python
for i, d in enumerate(data):  
    print(i, d)
# 이게 아니다     
```

```python
for d_index,d in enumerate(data):     #d =0(name),1(hobby)
    print("data",d_index,":", d)
    for i_index,i in enumerate(d):   #i=0,1,2,3,4,5  #for i in data[0]
        print(i_index, i)
        
    print('----------------------')
        
# '이원재' =data[0][0]        
# ★이중 for문은 인덱스 d, i 값이 중요하다.
# ★enumerate랑 같이 써서 확인하면 편함 
```

- dataframe도 이와 같다.

```python
df_1 = pd.DataFrame([[1, 2, 3], [4, 5, 6]], 
                    columns=['a', 'b', 'c'])
df_1
```

```python
#데이터 수정 : 'a' <- 
df_1['a'] = df_1['b']

#데이터 추가 : 'd'
df_1['d'] = df_1['c']
df_1
```

```python
df_1['a'] += 3
df_1
```

```python
df_1['a'] = df_1['b']*df_1['c']
df_1
```

```python
# columns 지정하는 또다른 방법
df_1.a = df_1.b * df_1.c
df_1
```

- 아래와 같이 숫자로 형변환이 안되는 경우

```python
df_2 = pd.DataFrame([['1,000', '2,000', '3,000'],
                     ['1', '2', '3']])
```

```python
df_2
```

```python
df_2.stack() 
```

```python
df_2 = df_2.stack().str.replace(",", "").unstack()
```

```python
df_2
```

```python
df_2.astype(float)
```

```python
df_2 = df_2.astype(int)
df_2
```

```python
multi_ind = pd.MultiIndex.from_tuples([('a', 0), ('a', 1)])
 
multi_col = [['b', 'b', 'b'], [0, 1, 2]]
```

```python
df_mul = pd.DataFrame(data=df_2, 
                      columns=multi_col, 
                      index=multi_ind)
df_mul
```
